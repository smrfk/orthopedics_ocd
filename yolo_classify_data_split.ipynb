# norm vs ocd

```python
# split dataset
import os
import random
import shutil

# ディレクトリのパス
top_folder = r"C:\Users\fukud\Documents\orthopedics\ultralytics-main\examples\classification_nor_vs_ocd"

# 分割する割合
train_ratio = 0.6
valid_ratio = 0.2
test_ratio = 0.2

# train, valid, testフォルダの作成
os.makedirs(os.path.join(top_folder, "train"), exist_ok=True)
os.makedirs(os.path.join(top_folder, "valid"), exist_ok=True)
os.makedirs(os.path.join(top_folder, "test"), exist_ok=True)

# normal, abnormalフォルダの処理
for label in ["normal", "ocd"]:
    # train, valid, testフォルダの作成
    os.makedirs(os.path.join(top_folder, "train", label), exist_ok=True)
    os.makedirs(os.path.join(top_folder, "valid", label), exist_ok=True)
    os.makedirs(os.path.join(top_folder, "test", label), exist_ok=True)

    # AL, AS, PL, PSフォルダの処理
    for folder_name in ["AL", "AS", "PL", "PS"]:
        folder_path = os.path.join(top_folder, label, folder_name)

        # ファイル名を取得してシャッフル
        files = os.listdir(folder_path)
        random.seed(1231)
        random.shuffle(files)

        # ファイル数と分割数の計算
        num_files = len(files)
        #print("num_files", num_files)
        num_train = int(num_files * train_ratio)
        #print("num_train", num_train)
        num_valid = int(num_files * valid_ratio)
        #print("num_valid", num_valid)
        num_test = num_files - num_train - num_valid
        #print("num_test", num_test)

        # trainフォルダにコピー
        for file_name in files[:num_train]:
            src_path = os.path.join(folder_path, file_name)
            dst_path = os.path.join(top_folder, "train", label, folder_name, file_name)
            dst_folder_path = os.path.dirname(dst_path)
            os.makedirs(dst_folder_path, exist_ok=True)
            shutil.copy(src_path, dst_path)

        # validフォルダにコピー
        for file_name in files[num_train:num_train+num_valid]:
            src_path = os.path.join(folder_path, file_name)
            dst_path = os.path.join(top_folder, "valid", label, folder_name, file_name)
            dst_folder_path = os.path.dirname(dst_path)
            os.makedirs(dst_folder_path, exist_ok=True)
            shutil.copy(src_path, dst_path)

        # testフォルダにコピー
        for file_name in files[num_train+num_valid:]:
            src_path = os.path.join(folder_path, file_name)
            dst_path = os.path.join(top_folder, "test", label, folder_name, file_name)
            dst_folder_path = os.path.dirname(dst_path)
            os.makedirs(dst_folder_path, exist_ok=True)
            shutil.copy(src_path, dst_path)
```

```python
# ルートフォルダのパス
root_folder_path = r"C:\Users\fukud\Documents\orthopedics\ultralytics-main\examples\classification_nor_vs_ocd"

# フォルダ名
folder_names = ["train", "valid", "test"]

# ルートフォルダ以下のフォルダを順番に処理
for folder_name in folder_names:
    folder_path = os.path.join(root_folder_path, folder_name)
    
    # AL, AS, PL, PSのフォルダを順番に処理
    for sub_folder_name in ["normal", "ocd"]:
        sub_folder_path = os.path.join(folder_path, sub_folder_name)
        
        # AL, AS, PL, PSのフォルダ以下のファイルを順番に処理
        for subsub_folder_name in os.listdir(sub_folder_path):
            subsub_folder_path = os.path.join(sub_folder_path, subsub_folder_name)

            for file_name in os.listdir(subsub_folder_path):
                file_path = os.path.join(subsub_folder_path, file_name)
            
                # normalまたはabnormalフォルダに移動する
                destination_folder_path = os.path.join(sub_folder_path, file_name)
                shutil.move(file_path, destination_folder_path)
```

```python
import os
import shutil
import random

# データセットのルートディレクトリ
root_dir = r'C:\Users\fukud\Documents\orthopedics\data\ocd_classify_4classes'

# train, val, testディレクトリを作成
os.makedirs(os.path.join(root_dir, 'train'), exist_ok=True)
os.makedirs(os.path.join(root_dir, 'val'), exist_ok=True)
os.makedirs(os.path.join(root_dir, 'test'), exist_ok=True)

# AS, AL, PS, PLディレクトリを作成
subdirs = ['AS', 'AL', 'PS', 'PL']
for subdir in subdirs:
    os.makedirs(os.path.join(root_dir, 'train', subdir), exist_ok=True)
    os.makedirs(os.path.join(root_dir, 'val', subdir), exist_ok=True)
    os.makedirs(os.path.join(root_dir, 'test', subdir), exist_ok=True)

# データセット内のnormalとabnormalディレクトリを取得
for subdir in subdirs:
    normal_dir = os.path.join(root_dir, subdir, 'normal')
    abnormal_dir = os.path.join(root_dir, subdir, 'ocd')

    # トレーニング、検証、テストセットの比率を設定
    train_ratio = 0.6
    val_ratio = 0.2
    test_ratio = 0.2

    # normalとabnormalディレクトリのファイルリストを取得
    normal_files = os.listdir(normal_dir)
    abnormal_files = os.listdir(abnormal_dir)

    # normalファイルのシャッフル
    random.shuffle(normal_files)

    # トレーニング、検証、テストセットのファイル数を計算
    num_normal = len(normal_files)
    num_train_normal = int(num_normal * train_ratio)
    num_val_normal = int(num_normal * val_ratio)
    num_test_normal = num_normal - num_train_normal - num_val_normal

    # トレーニングセットのnormalファイルをコピー
    for i in range(num_train_normal):
        src_file = os.path.join(normal_dir, normal_files[i])
        dst_file = os.path.join(root_dir, 'train', subdir, normal_files[i])
        shutil.copy(src_file, dst_file)

    # 検証セットのnormalファイルをコピー
    for i in range(num_train_normal, num_train_normal+num_val_normal):
        src_file = os.path.join(normal_dir, normal_files[i])
        dst_file = os.path.join(root_dir, 'val', subdir, normal_files[i])
        shutil.copy(src_file, dst_file)

    # テストセットのnormalファイルをコピー
    for i in range(num_train_normal+num_val_normal, num_normal):
        src_file = os.path.join(normal_dir, normal_files[i])
        dst_file = os.path.join(root_dir, 'test', subdir, normal_files[i])
        shutil.copy(src_file, dst_file)

    # abnormalファイルのシャッフル
    random.shuffle(abnormal_files)

    # トレーニング、検証、テストセットのファイル数を計算
    num_abnormal = len(abnormal_files)
    num_train_abnormal = int(num_abnormal * train_ratio)
    num_val_abnormal = int(num_abnormal * val_ratio)
    num_test_abnormal = num_abnormal - num_train_abnormal - num_val_abnormal

    # トレーニングセットのabnormalファイルをコピー
    for i in range(num_train_abnormal):
        src_file = os.path.join(abnormal_dir, abnormal_files[i])
        dst_file = os.path.join(root_dir, 'train', subdir, abnormal_files[i])
        shutil.copy(src_file, dst_file)

    # 検証セットのabnormalファイルをコピー
    for i in range(num_train_abnormal, num_train_abnormal+num_val_abnormal):
        src_file = os.path.join(abnormal_dir, abnormal_files[i])
        dst_file = os.path.join(root_dir, 'val', subdir, abnormal_files[i])
        shutil.copy(src_file, dst_file)

    # テストセットのabnormalファイルをコピー
    for i in range(num_train_abnormal+num_val_abnormal, num_abnormal):
        src_file = os.path.join(abnormal_dir, abnormal_files[i])
        dst_file = os.path.join(root_dir, 'test', subdir, abnormal_files[i])
        shutil.copy(src_file, dst_file)
    
```
